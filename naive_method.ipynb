```python
import numpy as np
```

```python
%cd /content/drive/MyDrive/
X_trn = np.loadtxt( "train_challenges.dat" )
%cd /content/drive/MyDrive/
y_trn = np.loadtxt( "train_responses.dat" )
```

```python
w = np.linalg.lstsq(X_trn , y_trn , rcond=None)[0]
print(len(w))
print (w) 
print (np.count_nonzero(w<0))
```

```python
def hardthresholding(w): 
    w_copy = np.copy(w);
    Kth_big = np.sort(w)[2048-512]
    i=0;
    while i<len(w):
      if(w_copy[i]<Kth_big):
        w_copy[i]=0;
      i=i+1;
    return w_copy;
```

```python
def mean_abs_error(y_trn,X_trn,w):
  i=0
  error = 0
  while i<len(y_trn):
    y_exp = np.dot(w.T,X_trn[i]);
    error=error+np.abs(y_trn[i]-y_exp)
    i=i+1
  return error/1600
```

```python
w_sp = hardthresholding(w)
print(np.count_nonzero(w_sp<0))
mean_error = mean_abs_error(y_trn,X_trn,w)
mean_error_1 = mean_abs_error(y_trn,X_trn,w_sp)
```
