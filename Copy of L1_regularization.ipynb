```python
p = np.linalg.lstsq(X_trn , Y_trn , rcond=None)[0]
reg_constant = 1
step = 0.01
```

```python
import numpy as np
```

```python
def grad_1(Y_trn,X_trn,p):             
  i=0
  sum_array = np.zeros(len(X_trn[0]))
  while i<len(X_trn):
    a = Y_trn[i] - np.dot(p,X_trn[i])
    b = np.negative(X_trn[i])
    c = np.multiply(b,a)
    sum_array = np.add(sum_array , c)
    i=i+1
  mean=np.divide(sum_array,len(X_trn))
  return mean

def grad_2 (reg_constant, p) :
  grad = np.zeros(len(p))
  i = len (p) - 1
  while (i>=0):
    if p[i]>0:
      grad [i] = reg_constant
    elif p[i]<0:
      grad [i] = -reg_constant
    else:
      grad[i] = 0
    i=i-1
  return grad

def grad_total(x,y):
  return np.add(x,y)
```

```python
def mymodel(X_trn,Y_trn,p):
  step1  = 0.1
  while(1):
     x = grad_1(Y_trn, X_trn, p)
     y = grad_2(reg_constant, p)
     grad_t = grad_total(x,y)
    
     t = np.subtract(p, np.multiply(step,grad_t))     #update the p vector

     error_1 = np.linalg.norm(np.subtract(t,p))
     if (error_1 < 30):                                #convergence criterion
        return t
     p = t
     step1 = step1
```

```python
def mean_abs_error(Y_trn,X_trn,t):
  i=0
  error = 0
  while i<len(Y_trn):
    y_exp = np.dot(t.T,X_trn[i]);
    error=error+np.abs((Y_trn[i]-y_exp))
    i=i+1
  return error/1600.0
```

```python
def hardthresholding(w): 
    w_copy = np.copy(w);
    Kth_big = np.sort(w)[2048-512]
    i=0;
    while i<len(w):
      if(w_copy[i]<Kth_big):
        w_copy[i]=0;
      i=i+1;
    return w_copy;
```

```python
t = mymodel(X_trn,Y_trn,p)
t_sparse = hardthresholding(t)
print(t_sparse)
print(mean_abs_error(Y_trn,X_trn,t))
print(mean_abs_error(Y_trn,X_trn,t_sparse))
```
