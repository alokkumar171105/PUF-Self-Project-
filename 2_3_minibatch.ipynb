```python
mini_batch = [200,400,600,800,1000,1200,1400]
print(len(mini_batch))
```

```python
import random
```

```python
regularization_constant = [0]*50
i=0
while (i<50):
  regularization_constant[i] = 0.1 + 0.1*i
  i=i+1
```

```python
p = np.linalg.lstsq(X_trn_1 , Y_trn_1 , rcond=None)[0]
step = 0.001
error = np.zeros((len(regularization_constant),7))
```

```python
import numpy as np
```

```python
def hardthresholding(w): 
    w_copy = w;
    Kth_big = np.sort(w)[2048-512]
    i=0;
    while i<len(w):
      if(w_copy[i]<Kth_big):
        w_copy[i]=0;
      i=i+1;
    return w_copy;
```

```python
def grad_1(Y_trn,X_trn,p):             
  i=0
  sum_array = [0.0]*(len(X_trn[0]))
  while i<len(X_trn):
    a = Y_trn[i] - np.dot(p,X_trn[i])
    b = np.negative(X_trn[i])
    c = np.multiply(b,a)
    sum_array = np.add(sum_array , c)
    i=i+1
  mean=np.divide(sum_array,len(X_trn))
  return mean

def grad_2 (reg_constant, p) :
  grad = [0.0]*(len(p))
  i = len (p) - 1
  while (i>=0):
    if p[i]>0:
      grad [i] = reg_constant
    else :
      grad [i] = -reg_constant
    i=i-1
  return grad

def grad_total(x,y):
  return np.add(x,y)
```

```python
def mymodel(X_trn,Y_trn,p,reg_constant,mini_batc):
  step1  = 0.1
  while(1):
     numbers = list(range(0, 1599))
     num_samples = mini_batc

     random_numbers = random.sample(numbers, num_samples)

     X_trn = X_trn_1[random_numbers]
     Y_trn = Y_trn_1[random_numbers]

     x = grad_1(Y_trn, X_trn, p)
     y = grad_2(reg_constant, p)
     grad_t = grad_total(x,y)
    
     t = np.subtract(p, np.multiply(step,grad_t))     #update the p vector
     t_sparse = hardthresholding(t)

     error_1 = np.linalg.norm(np.subtract(t_sparse,p))
     if (error_1 < 10):                                #convergence criterion
        return t
     p = t
```

```python
def mean_abs_error(Y_trn_1,X_trn_1,t):
  i=0
  error_2 = 0
  while i<len(Y_trn_1):
    y_exp = np.dot(t.T,X_trn_1[i]);
    error_2=error_2+np.abs((Y_trn_1[i]-y_exp))
    i=i+1
  return error_2/1600.0
```

```python
i=0
while(i<len(regularization_constant)):
  j=0
  r_constant = regularization_constant[i]
  while(j<len(mini_batch)):
    mini_bat = mini_batch[j]
    t = mymodel(X_trn_1,Y_trn_1,p,r_constant,mini_bat)
    t_sparse = hardthresholding(t)
    error[i][j] = mean_abs_error(Y_trn_1,X_trn_1,t_sparse)
    j=j+1
  i=i+1
```

```python
print(error)
min_index = np.unravel_index(np.argmin(error), error.shape)

print(min_index)
print(error[min_index])
```
